```python
import cv2
from PIL import Image
import numpy as np
import pathlib
import time
from ultralytics import YOLO  # Import YOLO class from ultralytics

# Pathfix
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath

# Model
# Make sure this path is correct and points to your YOLOv8 model
model = YOLO('C:/Users/lee/Desktop/knife_detection/best.pt')

#   
cap = cv2.VideoCapture(0)  #    0  

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # OpenCV  PIL  
    img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    
    #   
    results = model(img)  # Run the model on the image
    
    #    render() 
    results[0].render()  # Render the detections on the image
    
    #   OpenCV 
    cv2.imshow('YOLOv8 Detection', np.squeeze(results[0].imgs[0]))  # Show the result image

    try:
        # Access the detected objects using pandas (results.pandas() returns a DataFrame)
        if "knife" in results[0].pandas().xyxy[0]["name"].values or "guns" in results[0].pandas().xyxy[0]["name"].values:
            print("knife or guns detected")
        else:
            print("knife or guns not detected")
    except Exception:
        pass

    if cv2.waitKey(1) & 0xFF == ord('q'):  # q  
        break
    time.sleep(2)

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import math 
# start webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

# model
model = YOLO("C:/Users/lee/Desktop/knife_detection/best.pt")

# object classes
classNames = ["knife", "guns"
              ]


while True:
    success, img = cap.read()
    results = model(img, stream=True)

    # coordinates
    for r in results:
        boxes = r.boxes

        for box in boxes:
            # bounding box
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2) # convert to int values

            # put box in cam
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 3)

            # confidence
            confidence = math.ceil((box.conf[0]*100))/100
            print("Confidence --->",confidence)

            # class name
            cls = int(box.cls[0])
            print("Class name -->", classNames[cls])

            # object details
            org = [x1, y1]
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2

            cv2.putText(img, classNames[cls], org, font, fontScale, color, thickness)

    cv2.imshow('Webcam', img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
